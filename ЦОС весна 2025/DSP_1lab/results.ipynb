```python
import matplotlib.pyplot as plt
import numpy as np

#    
data = np.loadtxt('signal.txt')

#  
plt.figure(figsize=(10, 6))
plt.plot(data[:, 0], data[:, 1])
plt.xlabel('')
plt.ylabel('')
plt.title(' ')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#    
data = np.loadtxt('overlapped_signal.txt')

#  
plt.plot(data[:, 0], data[:, 1], label=' ')
plt.plot(data[:, 0], data[:, 2], label=' ')
plt.legend()
plt.xlabel('')
plt.ylabel('')
plt.title('')
plt.show()
```
