```python
import matplotlib.pyplot as plt
import numpy as np

# Чтение данных из файла
data = np.loadtxt('signal.txt')

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(data[:, 0], data[:, 1])
plt.xlabel('Время')
plt.ylabel('Амплитуда')
plt.title('График сигнала')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Чтение данных из файла
data = np.loadtxt('overlapped_signal.txt')

# Построение графика
plt.plot(data[:, 0], data[:, 1], label='Исходный сигнал')
plt.plot(data[:, 0], data[:, 2], label='Квантованный сигнал')
plt.legend()
plt.xlabel('Время')
plt.ylabel('Значение')
plt.title('Сигналы')
plt.show()
```
